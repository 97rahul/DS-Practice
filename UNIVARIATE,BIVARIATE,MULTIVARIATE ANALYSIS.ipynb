```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset("iris")
df
```

# UNIVARIATE ANALYSIS

```python
setosa = df.loc[df['species']=='setosa']
virginica = df.loc[df['species']=='virginica']
versicolor = df.loc[df['species']=='versicolor']
```

```python
plt.plot(setosa.sepal_length,np.zeros_like(setosa.sepal_length),'*')
plt.plot(virginica.sepal_length,np.zeros_like(virginica.sepal_length),'*')
plt.plot(versicolor.sepal_length,np.zeros_like(versicolor.sepal_length),'*')
plt.xlabel("Sepal Length")
plt.show()
```

OBSERVATION:
    A line cluster for classification of species based on sepal length but points gets overlapped.
    Hence ambiguity.

# BIVARIATE ANALYSIS

```python
sns.FacetGrid(df,hue="species",size=5).map(plt.scatter,'petal_length','sepal_width').add_legend()
```

OBSERVATION:
    Clusters are formed for each species but there is an overlapping of versicolor and verginica species.
    Hence not a clear boundary between the two.

# MULTIVARIATE ANALYSIS

```python
sns.pairplot(df,hue="species",size=3)
```

OBSERVATION:
    Clear distinction for setosa species but yet not betweeen virginica and versicolor.
    Gives a clear picture on features linear dependency.
    Helps to find correlation and covariance among features.
