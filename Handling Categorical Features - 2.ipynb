## 4.Target Guided Ordinal Encoding 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv', usecols=['Cabin','Survived'])
df.head()
```

```python
df["Cabin"].fillna("Missing",inplace = True)
df
```

```python
df["Cabin"] = df["Cabin"].astype(str).str[0]
df
```

```python
df.Cabin.unique()
```

```python
o_labels = df.groupby(["Cabin"])["Survived"].mean().sort_values().index
o_labels
```

```python
o_labels2 = {k:i for i,k in enumerate(o_labels,0)}
o_labels2
```

```python
df['Cabin_ordinal_labels'] = df['Cabin'].map(o_labels2)
df.head()
```

## 4. Mean Encoding

```python
o_labels = df.groupby(["Cabin"])["Survived"].mean()
o_labels = o_labels.to_dict()
o_labels
```

```python
df['mean_ordinal_encode']=df['Cabin'].map(o_labels)
df.head()
```

## 5. Probability Ratio Encoding

```python
df = pd.read_csv('titanic.csv', usecols=['Cabin','Survived'])
df.head()
```

```python
df["Cabin"].fillna("Missing",inplace = True)
df
```

```python
df["Cabin"] = df["Cabin"].astype(str).str[0]
df
```

```python
df.Cabin.unique()
```

```python
prob_df = df.groupby(["Cabin"])["Survived"].mean()
```

```python
prob_df = pd.DataFrame(prob_df,columns = ["Survived"])
prob_df
```

```python
prob_df["Dead"] = 1-prob_df["Survived"]
prob_df
```

```python
prob_df['Probability_ratio']=prob_df['Survived']/prob_df['Dead']
prob_df.head()
```

```python
p_encoded = prob_df['Probability_ratio'].to_dict()
p_encoded
```

```python
df['Cabin_encoded']=df['Cabin'].map(p_encoded)
df.head()
```
