# VAEX

```python
import vaex
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randint(0,100,size = (1000000,500)),columns = ['col%d' % i for i in range(500)])
df.head()
```

```python
df.info(memory_usage='deep')
```

```python
df.to_csv('sample.csv',index = False)
```

```python
vaex_df = vaex.from_csv('sample.csv', convert=True, chunk_size = 5_000_000)
type(vaex_df)
```

```python
type(vaex_df)
```

```python
vaex_df = vaex.open('sample.csv.hdf5')
```
